# Gymnasium
## Использование среды Cart Pole

### 1. Импорт библиотеки:

```python
import gymnasium as gym
import time
```

gymnasium: Библиотека для создания и работы с различными симуляционными средами для задач обучения с подкреплением.

time: Используется для добавления задержки между шагами симуляции.

### 2. Создание среды:

```python
env = gym.make('CartPole-v1', render_mode="human")
```

gym.make: Создаёт среду по имени. В данном случае это CartPole-v1, классическая задача из обучения с подкреплением.

render_mode="human": Включает визуализацию среды в режиме реального времени.

### 3. Инициализация среды:

```python
obs, info = env.reset()
```

env.reset(): Сбрасывает состояние среды в начальное, чтобы начать новый эпизод.

obs: Наблюдение за состоянием среды, в данном случае это массив с положением тележки, углом шеста и другими параметрами.

info: Дополнительная информация о среде.

### 4. Основной цикл симуляции:

```python
for _ in range(500):  # 500 шагов
    action = env.action_space.sample()  # Случайное действие
    obs, reward, done, truncated, info = env.step(action)

    # Обновление окна для обработки событий
    env.render()

    # Добавляем небольшую задержку
    time.sleep(0.05)

    # Проверяем завершение эпизода
    if done or truncated:
        print("Эпизод завершён! Падение:", done)
        obs, info = env.reset()
env.close()
```

Выполняется 500 шагов симуляции или до завершения эпизода.

env.action_space.sample(): Случайное действие, допустимое в данной среде. В CartPole-v1 возможны два действия: толкнуть тележку влево или вправо.

env.step(action): Выполняет одно действие в среде и возвращает:

obs: Новое состояние среды.

reward: Награда за текущее действие.

done: Флаг завершения эпизода (например, если шест упал).

truncated: Флаг, указывающий, было ли выполнение прервано по ограничениям среды (например, превышено максимальное количество шагов).

info: Дополнительная информация о текущем состоянии.

env.render(): Обновляет визуализацию текущего состояния среды.

time.sleep(0.01): Добавляет задержку в 50 мс для плавности анимации.

Если эпизод завершён (done=True) или был принудительно прерван (truncated=True), среда сбрасывается.

env.close(): Закрывает визуализацию и освобождает ресурсы.
