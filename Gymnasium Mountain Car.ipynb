# Gymnasium
## Использование Среды Mountain Car

### 1. Импорт библиотеки:

```python
import gymnasium as gym
```

gymnasium: Библиотека для создания и работы с различными симуляционными средами для задач обучения с подкреплением.

### 2. Создание среды:

```python
env = gym.make('MountainCar-v0', render_mode="human")
```

gym.make('MountainCar-v0'): Создаёт среду MountainCar-v0. В данной задаче маленькая машинка должна преодолеть подъем.

render_mode="human": Включает визуализацию в реальном времени.

### 3. Инициализация среды:

```python
obs, info = env.reset()
```

env.reset(): Сбрасывает состояние среды для начала нового эпизода.

obs: Текущее состояние среды (вектор с двумя значениями):

obs[0]: Положение машинки.

obs[1]: Скорость машинки.

info: Дополнительная информация о среде.

### 4. Основной цикл симуляции:

```python
for _ in range(1000):
    action = 2
    
    obs, reward, done, truncated, info = env.step(action)
    
    if done:
       print("Игра завершена!")
       obs, info = env.reset()
env.close()
```

Этот цикл выполняет 1000 итераций взаимодействия с окружением:

action = 2: Здесь фиксировано одно действие — "ускорить вправо". В MountainCar-v0 доступны три действия:

0: Ускорение влево.

1: Никакого ускорения.

2: Ускорение вправо.

env.step(action): Выполняет указанное действие и возвращает:

obs: Новое наблюдение (состояние среды).

reward: Награда за текущее действие. В MountainCar-v0 обычно фиксирована на значении -1 за каждый шаг, пока цель не достигнута.

done: Булевый флаг. Показывает, достигнуто ли конечное состояние (машинка перевалилась через гору).

truncated: Булевый флаг. Показывает, завершилась ли игра из-за ограничения по времени.

info: Дополнительная информация.

Если машинка достигает цели (переваливается через вершину), устанавливается done = True, и игра считается завершенной.

После этого среда сбрасывается в начальное состояние для новой игры.

env.close() : завершения работы программа закрывает графическое окно, чтобы освободить ресурсы.
